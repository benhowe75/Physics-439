```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
# constants
g = 9.8
lh = 0.1
N = 5000
h = .001
```

```python
theta =np.zeros(N+1)
thetadot = np.zeros(N+1)
t = np.zeros(N+1)
```

```python
def omega(thetadot):
    return thetadot
def omegadot(theta):
    return g / lh * np.sin(theta)
```

```python
theta[0] = 0.3
thetadot[0] = 0
t[0] = 0
```

```python
for i in range(N):
    k1_theta = h * omega(thetadot[i])
    k1_thetadot = h * omegadot(theta[i])
    
    k2_theta = h * omega(thetadot[i] + k1_thetadot/2.0)
    k2_thetadot = h * omegadot(theta[i] + k1_theta/2.0)
    
    k3_theta = h * omega(thetadot[i] + k2_thetadot/2.0)
    k3_thetadot = h * omegadot(theta[i] + k2_theta/2.0)
    
    k4_theta = h * omega(thetadot[i] + k3_thetadot)
    k4_thetadot = h * omegadot(theta[i] + k3_theta)
    
    theta[i+1] = theta[i] + (k1_theta + 2.0 * k2_theta + 2.0 * k3_theta + k4_theta)/6.0
    thetadot[i+1] = thetadot[i] + (k1_thetadot + 2.0 * k2_thetadot + 2.0 * k3_thetadot + k4_thetadot)/6.0
    t[i+1] = t[i] + h
```

```python
plt.plot(t,theta)
plt.grid(b=True, which='both', color='0.65', linestyle='-')
plt.xlabel('t (s)', fontsize=12)
plt.ylabel('\u03b8 (rad)', fontsize=12)
```
